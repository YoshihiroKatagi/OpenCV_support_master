```python
import cv2
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use("ggplot")
```

```python
# 日本語対応のフォントへ変更
# 入っていない場合は、自分で追加する必要あり
mpl.rcParams["font.family"] = "IPAGothic"
```

```python
img = cv2.imread("src/Lena.jpg",0)
#ヒストグラムの計算
hist = cv2.calcHist([img],[0],None,[256],[0,256])
plt.plot(hist)
plt.xlabel("画素値")
plt.ylabel("頻度")
```

```python
# ヒストグラム均一化
img_eq = cv2.equalizeHist(img)
hist_eq = cv2.calcHist([img_eq],[0],None,[256],[0,256])
plt.plot(hist_eq)
plt.xlabel("画素値")
plt.ylabel("頻度")
```

```python
# オリジナルの画像
cv2.imshow("img", img)
# ヒストグラム均一化後の画像
cv2.imshow("equalize_img", img_eq)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
